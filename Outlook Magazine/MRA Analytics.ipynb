### Getting Dummies for Categorical Variables

```python
def dummies_and_drop(df,key):
    return pd.concat([pd.get_dummies(df[key], prefix=key), df.drop([key], axis=1)], axis=1)
    
```

```python
pd.get_dummies(cars_data['Model']).head()
```

```python
cars_data = dummies_and_drop(cars_data, 'Model')
```

```python
pd.get_dummies(cars_data['Doors'], prefix='doors')
```

```python
cars_data = dummies_and_drop(cars_data, 'Doors')
```

```python
pd.get_dummies(cars_data['HP'])
```

```python
cars_data = dummies_and_drop(cars_data, 'HP')
```

### Training Model

```python
from sklearn import linear_model
from sklearn import metrics
from sklearn.metrics import mean_squared_error as MSE
from sklearn.model_selection import train_test_split
```

```python
reg = linear_model.LinearRegression()
```

##### Train Test Split

```python
y = cars_data['Price'].values
x = cars_data.drop(['Price'], axis=1)
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.2)
```

```python
test_x.columns
```

```python
reg.fit(train_x, train_y)
```

```python
plt.figure(figsize=(60,40))
plt.bar(range(len(reg.coef_)), reg.coef_)
```

```python
feature_importances = dict(zip([str(x) for x in list(train_x.columns)], list(reg.coef_)))
```

```python
feature_importances_df = pd.DataFrame(zip([str(x) for x in list(train_x.columns)], list(reg.coef_)))
```
