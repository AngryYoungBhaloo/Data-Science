##### Removing Super Overs from Match

```python
#because super over stats aren't counted

def remove_super_overs(match):
    match_without_super_over = match.loc[match["is_super_over"] == 0] #if the value is 0, that means it isn't a super over
    match = match_without_super_over
    return match
```

##### Wide Balls played by each player

```python
#the balls data counts the wide balls in the batsman's balls. These must be subtracted

def extra_balls(match):
    extra_balls = match[['batsman','wide_runs','noball_runs']] #only the relevant columns out of the df are extracted
    extra_balls.loc[extra_balls.wide_runs > 0, 'wide_runs'] = 1
    extra_balls.loc[extra_balls.noball_runs > 0, 'noball_runs'] = 1
    extra_balls["extra_balls"] = extra_balls['wide_runs'] + extra_balls['noball_runs']

    extra_balls_played = extra_balls.loc[:, ["batsman","extra_balls"]].groupby(["batsman"]).sum()
    
    return extra_balls_played
```

##### 4's and 6's

```python
def fours(match):
    only_runs = match[['batsman','batsman_runs']] #only the relevant columns out of the df are extracted
    
    only_runs.loc[only_runs['batsman_runs'] != 4, 'batsman_runs'] = 0 #making every other score expect 4 runs zero
    
    only_runs.loc[only_runs['batsman_runs'] == 4, 'batsman_runs'] = 1 #all the 4's
    
    fours_hit = only_runs.groupby(['batsman']).sum() #groupby batsman and count the 4's
    
    fours_hit.columns = ['fours_hit']
    
    return fours_hit
```

```python
def sixes(match):
    only_runs = match[['batsman','batsman_runs']] #only the relevant columns out of the df are extracted
    
    only_runs.loc[only_runs['batsman_runs'] != 6, 'batsman_runs'] = 0 #making every other score expect 6 runs zero
    
    only_runs.loc[only_runs['batsman_runs'] == 6, 'batsman_runs'] = 1 #all the 6's
    
    sixes_hit = only_runs.groupby(['batsman']).sum() #groupby batsman and count the 6's
    
    sixes_hit.columns = ['sixes_hit']
    
    return sixes_hit
```

#### Subtracting Byes and Leg Byes

```python
def subtract_byes(match_runs_in_overs):
    #adding byes and leg byes
    match_runs_in_overs['all_bye_runs'] = match_runs_in_overs.bye_runs + match_runs_in_overs.legbye_runs 
    #byes aren't added to a bowlers runs
    match_runs_in_overs['runs_conceded_by_bowler'] = match_runs_in_overs.total_runs - match_runs_in_overs.all_bye_runs
    
    #we don't need the other columns
    match_runs_in_overs = match_runs_in_overs.drop(['bye_runs', 'legbye_runs', 'total_runs', 'all_bye_runs'], axis=1) 
    
    return match_runs_in_overs
```

#### Overs Bowled

```python
# find the number of bowls bowled and then find the no. of wides and no balls etc. 
# subtract them to get bowls and divide to get overs

def bowls_bowled(match):
    #the counts give the no. of instances a bowler bowls, 
    match_total_balls = match.groupby('bowler').count()[['ball']]
    
    #getting instances where wide runs and no ball runs are not zero. only extracting three columns from this dataframe
    match_illegal_deliveries = match.loc[(match['wide_runs'] != 0) | (match['noball_runs'] != 0), ['bowler','wide_runs', 'noball_runs']]
    
    #as wide runs and no ball runs cannot be both awarded on the same bowl, the new column formed by adding them
    #will only have a value because of one or the other. Then a count of all these instances will give use all instances
    #of illegal deliveries
    match_illegal_deliveries['wide_and_no_ball_runs'] = match_illegal_deliveries.wide_runs + match_illegal_deliveries.noball_runs
    
    match_illegal_deliveries = match_illegal_deliveries.drop(['wide_runs', 'noball_runs'], axis=1)
    #all these records are the ones where a wide ball or a no ball was bowled. a count will give all illegal deliveries
    match_illegal_deliveries = match_illegal_deliveries.groupby(['bowler']).count()
    #renaming the column 
    match_illegal_deliveries.columns = ['illegal_deliveries']
    
    #concatenating to get the illegal deliveries alongside bowler name in the main dataframe
    match_total_balls = pd.concat([match_total_balls, match_illegal_deliveries], axis=1)
    
    #some of the bowlers might not have bowled an illegal delivery and so there entries appear as nan
    match_total_balls = match_total_balls.fillna(0)
    
    match_total_balls['legal_deliveries_bowled'] = match_total_balls.ball - match_total_balls.illegal_deliveries
    
    match_total_balls = match_total_balls.drop(['ball', 'illegal_deliveries'], axis=1)
    
    return match_total_balls
```

### Wickets

```python
def wickets(match):
    match_wickets_by_bowlers = match[['bowler', 'dismissal_kind']] #we only need these columns
    
    #the modes of dismissals which do not go to the bowler are replaced with a zero
    match_wickets_by_bowlers.loc[match_wickets_by_bowlers.dismissal_kind.isin(bowler_dismissals) == False, 'dismissal_kind'] = 0
    
    #modes of dismissals which go to the bowler are replaced with one
    #this way I can sum these to get the wickets for every bowler
    match_wickets_by_bowlers.loc[match_wickets_by_bowlers.dismissal_kind.isin(bowler_dismissals), 'dismissal_kind'] = 1
    
    match_wickets_by_bowlers = match_wickets_by_bowlers.groupby('bowler').sum()
    
    match_wickets_by_bowlers.columns = ['wickets'] #change column name
    
    return match_wickets_by_bowlers
```

#### Economy

```python
def bowler_economy(match_runs_by_bowlers, match_total_balls):
    match_bowler_economy = pd.concat([match_runs_by_bowlers, match_total_balls], axis=1)
    
    match_bowler_economy['economy'] = match_bowler_economy.runs_conceded_by_bowler/match_bowler_economy.legal_deliveries_bowled*6
    
    return match_bowler_economy
```

##### Economy Bonus

```python
def economy_bonus(match_bowler_economy):
    match_bowler_economy['economy_bonus'] = 0
    
    match_bowler_economy.loc[(match_bowler_economy.legal_deliveries_bowled > 12) & (match_bowler_economy.economy < 5), 'economy_bonus'] = 1
    match_bowler_economy.loc[(match_bowler_economy.legal_deliveries_bowled > 12) & (match_bowler_economy.economy >= 5) & (match_bowler_economy.economy < 6), 'economy_bonus'] = 2
    match_bowler_economy.loc[(match_bowler_economy.legal_deliveries_bowled > 12) & (match_bowler_economy.economy >= 6) & (match_bowler_economy.economy <= 7), 'economy_bonus'] = 3
    
    match_bowler_economy.loc[(match_bowler_economy.legal_deliveries_bowled > 12) & (match_bowler_economy.economy >= 10) & (match_bowler_economy.economy <= 11), 'economy_bonus'] = 4
    match_bowler_economy.loc[(match_bowler_economy.legal_deliveries_bowled > 12) & (match_bowler_economy.economy > 11) & (match_bowler_economy.economy <= 12), 'economy_bonus'] = 5
    match_bowler_economy.loc[(match_bowler_economy.legal_deliveries_bowled > 12) & (match_bowler_economy.economy > 12), 'economy_bonus'] = 6
    
    return match_bowler_economy
```

### Bowler Match Stats

```python
def get_bowlers_data(match):
    match_runs_in_overs = runs_in_overs(match) 
    
    add_bowlers_bowling_overs(match_runs_in_overs)
    
    match_runs_in_overs = subtract_byes(match_runs_in_overs)
    
    maiden_or_not(match_runs_in_overs)
    
    match_runs_by_bowlers = runs_by_bowlers(match_runs_in_overs)
    
    match_total_balls = bowls_bowled(match)
    
    match_wickets_by_bowlers = wickets(match)
    
    match_wickets_by_bowlers['wicket_landmark'] = 0
    match_wickets_by_bowlers.loc[match_wickets_by_bowlers.wickets == 3, 'wicket_landmark'] = 1
    match_wickets_by_bowlers.loc[match_wickets_by_bowlers.wickets == 4, 'wicket_landmark'] = 2
    match_wickets_by_bowlers.loc[match_wickets_by_bowlers.wickets == 5, 'wicket_landmark'] = 3
    
    match_lbw_and_bowled = lbw_and_bowled(match)
    
    match_bowler_economy = bowler_economy(match_runs_by_bowlers, match_total_balls)
    
    match_bowler_economy = economy_bonus(match_bowler_economy)
    
    match_bowler_data = pd.concat([match_wickets_by_bowlers, match_bowler_economy, match_lbw_and_bowled], axis=1)
      
    return match_bowler_data
```

#### Run outs

```python
def run_outs(match):
    run_outs = match[['dismissal_kind', 'fielder']]
    
    run_outs = run_outs.loc[run_outs.dismissal_kind == 'run out']
    
    match_run_outs = run_outs.groupby(['fielder']).count()
    
    match_run_outs.columns = ['run_outs']
    return match_run_outs
```

#### Stumpings

```python
def stumpings(match):
    stumpings = match[['dismissal_kind', 'fielder']]
    stumpings = stumpings.loc[stumpings.dismissal_kind == 'stumped']
    
    match_stumpings = stumpings.groupby(['fielder']).sum()
    
    match_stumpings.columns = ['stumpings']
    return match_stumpings
```

```python
def get_fielding_data(match):
    match_catches = catches(match)
    match_catches = catches_milestone(match_catches)
    
    match_run_outs = run_outs(match)
    
    match_stumpings = stumpings(match)
    
    match_fielder_data = pd.concat([match_catches, match_run_outs, match_stumpings], axis=1)
    
    match_fielder_data = match_fielder_data.fillna(0)
    
    return match_fielder_data
```
